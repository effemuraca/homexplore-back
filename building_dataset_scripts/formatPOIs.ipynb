```python
input_folder = '/drive/MyDrive/Intermediate Corporation/Dati/OpenStreetMap'
import pandas as pd
import os
# compact all the .csv files in one unique file called pois.csv
def compact_pois():
    pois = pd.DataFrame()
    for file in os.listdir(input_folder):
        if file.endswith('.csv'):
            df = pd.read_csv(os.path.join(input_folder, file))
            pois = pd.concat([pois, df])
    pois.to_csv(os.path.join(input_folder, 'pois.csv'), index=False)

compact_pois()
```
